```python
#call in panda definitions
import pandas as pd

#Load files
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

#Read the school and the student files and store in pandas dataframes
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#Combine the two dataframes
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

District Summary

```python
#Calculate the total number of schools
total_num_schools = school_data_complete["school_name"].value_counts().count()
```

```python
# total number of students 
total_students = school_data_complete["student_name"].count()
```

```python
# total budget
total_budget = school_data["budget"].sum()
```

```python
# average math score
avg_math_score = school_data_complete["math_score"].sum()/total_students
```

```python
# average reading score
avg_reading_score = school_data_complete["reading_score"].sum()/total_students
```

```python
# percentage of math passing rate
percentage_math_passing = (school_data_complete.math_score[(school_data_complete.math_score > 69)].count())/total_students*100
# percentage of Reading passing rate
percentage_reading_passing = (school_data_complete.reading_score[(school_data_complete.reading_score > 69)].count())/total_students*100

# percentage of over all passing rate
percentage_overall_passing = (percentage_math_passing + percentage_reading_passing)/2
```

```python
#Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Number of Schools": [total_num_schools], 
                                 "Number of Students": [total_students], 
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [avg_math_score], 
                                 "Average Reading Score": [avg_reading_score],
                                 "% Passing Math": [percentage_math_passing],
                                 "% Passing Reading": [percentage_reading_passing],
                                 "% Overall Passing Rate": [percentage_overall_passing]})

district_summary = district_summary[["Number of Schools", "Number of Students", "Total Budget",
                                     "Average Math Score", 
                                     "Average Reading Score",
                                     "% Passing Math",
                                     "% Passing Reading",
                                     "% Overall Passing Rate"]]

# Display the data frame
district_summary
```

## School Summary

```python
# Determine the School Type and per school student count, and budget
school_types = school_data.set_index(["school_name"],["type"])
school_types
```

```python
# Calculate the average test scores
per_school_math = school_data_complete.groupby(["school_name"]).mean()["math_score"]
per_school_reading = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
per_school_reading
```

## Math Scores by Grade

```python
# Create data series of scores by grade levels using conditionals
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group each by school name
ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]

# Combine series into single data frame
scores_by_grade = pd.DataFrame({"9th": ninth_graders_scores, "10th": tenth_graders_scores,
                                "11th": eleventh_graders_scores, "12th": twelfth_graders_scores})

# Minor data munging
scores_by_grade = scores_by_grade[["9th", "10th", "11th", "12th"]]
scores_by_grade.index.name = None

# Display the data frame
scores_by_grade
```
